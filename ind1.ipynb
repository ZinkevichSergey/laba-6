## Задача: 

Грузоподъёмность Барона m1=30кг
Грузоподъёмность Арсена m2=40кг
Вес 1 патрона m3=0.03кг
Количество патрон в обойме n=30 
Сколько кг металлолома смогут унести вдвоём Арсен и Барон при меткости деда k=0.5
m=m3*n*k=0.03*30*0.5=0.45
Ответ:0.45кг
